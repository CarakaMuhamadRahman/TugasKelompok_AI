```python
# Import Library
import numpy as np
import pandas as pd
import joblib
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('golf_df.csv')
df.info()
df.head(11)
```

```python
# Change Windy Datatype
df_windy = df['Windy'].astype(str)

df = df.drop(['Windy'], axis=1)
df['Windy'] = df_windy

df.info()
```

```python
# One-Hot-Encoding
new_df = pd.get_dummies(df[['Outlook', 'Temperature', 'Humidity', 'Windy']])
new_df.head(5)
```

```python
new_df['Play'] = df['Play']
new_df = pd.DataFrame(new_df)
new_df.info()
new_df.head(11)
```

```python
X = new_df[['Outlook_overcast','Outlook_rainy','Outlook_sunny','Temperature_cool','Temperature_hot','Temperature_mild','Humidity_high','Humidity_normal','Windy_False','Windy_True']]
y = new_df['Play']


# Train Test Split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.5, random_state=123)

y_train
```

```python
# Train Model / classifier
from sklearn.tree import DecisionTreeClassifier

cls = DecisionTreeClassifier()


DT_Model = cls.fit(X_train, y_train)

# Evaluate Model
from sklearn.metrics import accuracy_score

y_pred = DT_Model.predict(X_test)

acc_store = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy : ',acc_store)
```

```python
# Predict ML Model / Classifier with the new Data
print(DT_Model.predict([[0,0,1,0,0,1,1,0,1,0]])[0])
```

```python
import pickle

pickle.dump(DT_Model, open("model.pkl","wb"))
```
